```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv('/Users/alex/Downloads/pullout/DS4300_data_2010.csv')
```

```python
df
```

```python
df1 = df.sample(frac=1)
```

```python
df1
```

```python
tics = list(df1['tic'].unique())
```

```python
tics
```

```python
formatted = {}
for tic in tqdm(tics):
    temp = {"tic":tic}
    ts_data = []
    df2 = df1[df1['tic'] == tic]
    statics = False
    for index, row in df2.iterrows():
        st = {}
        if not statics:
            #tic = tic
            conm = row['conm']
            exchg = row['exchg']
            cik = row['cik']
            add1 = row['add1']
            spcindcd = row['spcindcd']
            weburl = row['weburl']
            st = {"tic": tic, "conm": conm, "exchg": exchg, "cik": cik, "add1": add1, "spcindcd": spcindcd, "weburl": weburl}
            
            statics = True
            temp["static"]=st
        
        date = str(row['datadate'])
        div = row['div']
        ajexdi = row['ajexdi']
        cshoc = row['cshoc']
        cshtrd = row['cshtrd']
        eps = row['eps']
        prccd = row['prccd']
        prchd = row['prchd']
        prcld = row['prcld']
        prcod = row['prcod']
        
        ts = {"div": div, "ajexdi": ajexdi, "cshoc": cshoc, "cshtrd": cshtrd, "eps": eps, "prccd": prccd, "prchd": prchd, "prcld": prcld, "prcod": prcod}
        
        ts_data+=[{"date":date,"data":ts}]
        
    temp["timeseries"] = ts_data
    formatted[tic] = temp
    
```

```python
use_me_df = df1.set_index(['tic', 'datadate'])
timeseries = ["div", "ajexdi", "cshoc", "cshtrd", "eps", "prccd", "prchd", "prcld", "prcod"]
static = ['cusip', 'cik', 'conm', 'exchg', 'add1', 'spcindcd', 'weburl']
```

```python

formatted = []
for ticker in tqdm(tics):
    data_tick = use_me_df.xs(ticker)
    static_df = use_me_df.iloc[0].to_frame().loc[static]
    ticker_dict = list(static_df.to_dict().values())[0]
    ticker_dict['ticker'] = static_df.columns[0][0]

    ticker_dict['timeseries'] = list(data_tick[timeseries].reset_index().to_dict('index').values())

    formatted.append(ticker_dict)

```
