```python
import pandas as pd
```

```python
print("begin read")
df = pd.read_csv('DS4300_data_2010.csv')
print("finished read")
print("goodbye")
```

```python
df
```

```python
df_small = df.sample(frac=0.00001)
df_small
```

```python
asset_list = []
i = 0
while i < len(df_small):
    keys = []
    for asset in asset_list:
        for key in asset.keys():
            keys.append(key)
    cusip = str(df_small.iloc[i]['cusip'])
    if cusip in keys:
        print("found cusip")
        match = -1
        ii = 0
        for asset in asset_list:
            for key in asset.keys():
                if key == cusip:
                    match = ii
                ii += 1
        date = str(df_small.iloc[i]['datadate'])
        div = df_small.iloc[i]['div']
        ajexdi = df_small.iloc[i]['ajexdi']
        cshoc = df_small.iloc[i]['cshoc']
        cshtrd = df_small.iloc[i]['cshtrd']
        eps = df_small.iloc[i]['eps']
        prccd = df_small.iloc[i]['prccd']
        prchd = df_small.iloc[i]['prchd']
        prcld = df_small.iloc[i]['prcld']
        prcod = df_small.iloc[i]['prcod']
        ts = {"div": div, "ajexdi": ajexdi, "cshoc": cshoc, "cshtrd": cshtrd, "eps": eps, "prccd": prccd, "prchd": prchd, "prcld": prcld, "prcod": prcod}
        asset_list[match][cusip]["timeseries"][date] = ts
    else:
        print("adding cusip")
        tic = df_small.iloc[i]['tic']
        conm = df_small.iloc[i]['conm']
        exchg = df_small.iloc[i]['exchg']
        cik = df_small.iloc[i]['cik']
        add1 = df_small.iloc[i]['add1']
        spcindcd = df_small.iloc[i]['spcindcd']
        weburl = df_small.iloc[i]['weburl']
        st = {"tic": tic, "conm": conm, "exchg": exchg, "cik": cik, "add1": add1, "spcindcd": spcindcd, "weburl": weburl}
        
        date = str(df_small.iloc[i]['datadate'])
        div = df_small.iloc[i]['div']
        ajexdi = df_small.iloc[i]['ajexdi']
        cshoc = df_small.iloc[i]['cshoc']
        cshtrd = df_small.iloc[i]['cshtrd']
        eps = df_small.iloc[i]['eps']
        prccd = df_small.iloc[i]['prccd']
        prchd = df_small.iloc[i]['prchd']
        prcld = df_small.iloc[i]['prcld']
        prcod = df_small.iloc[i]['prcod']
        ts = {"div": div, "ajexdi": ajexdi, "cshoc": cshoc, "cshtrd": cshtrd, "eps": eps, "prccd": prccd, "prchd": prchd, "prcld": prcld, "prcod": prcod}
        
        new_asset = {cusip: {"static":st, "timeseries":{date: ts}}}
        asset_list.append(new_asset)
    i += 1

asset_list
```

```python
from pymongo import MongoClient
client = MongoClient()
db = client.test_database
import datetime
posts = db.posts
```

```python
result = posts.insert_many(asset_list)
result.inserted_ids
```
